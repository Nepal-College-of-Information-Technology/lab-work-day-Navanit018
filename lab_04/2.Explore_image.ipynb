```python
import cv2
import numpy as np

# Read JPG image
image = cv2.imread("person.jpg")

# Check if image is loaded
if image is None:
    print("Error: Image not found")
else:
    print("Image loaded successfully")
```

```python
# Display image information
height, width, channels = image.shape
print("Height:", height)
print("Width:", width)
print("Channels:", channels)
print("Data type:", image.dtype)
```

```python
# Save image in BMP format
cv2.imwrite("person.bmp", image)

print("JPG image successfully converted to BMP format")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("person.jpg")

# Print image properties
print("Image Shape (Height, Width, Channels):", image.shape)
print("Image Data Type:", image.dtype)

# Display pixel value at a specific location
x, y = 130, 140
for a in range(30):
    for b in range(30):
        print(f"Pixel value at ({x+a},{y+b}):", image[x+a, y+b])

# print(f"Pixel value at ({x},{y}):", image[x, y])
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


# Display image
plt.imshow(image_rgb)
plt.title("Original Image")
# plt.axis("off")
plt.show()
```
