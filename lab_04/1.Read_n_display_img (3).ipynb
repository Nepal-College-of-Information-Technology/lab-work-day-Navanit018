```python
from PIL import Image
import matplotlib.pyplot as plt

# Read image
image = Image.open("person.jpg").convert("RGB")

# Display image
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
import numpy as np

# Convert PIL Image  NumPy array
image_np = np.array(image)

# Print image properties
print("Image Shape (Height, Width, Channels):", image_np.shape)
print("Image Data Type:", image_np.dtype)

# Display pixel value at a specific location
x, y = 100, 100
print(f"Pixel value at ({x},{y}):", image_np[x, y])
```
